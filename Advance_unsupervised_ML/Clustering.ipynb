```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# generate sample data
X,y_tune = make_blobs(n_samples=300,centers=4,cluster_std=0.6,random_state=0)

# instantiate a Kmeans object with the number of clusters
kmeans = KMeans(n_clusters=4)

# fit the data to the Kmeans algorithm
kmeans.fit(X)

# get the predicted cluster labels for each data point
labels = kmeans.labels_
# plot the data points with different colors for the differnt clusters

plt.figure(figsize=(10,8))
plt.scatter(X[:,0],X[:,1],c=labels)
plt.title('K-Means Clustering')
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('diamonds')

# generate sample data
X = df[['carat','price']]


# instantiate a Kmeans object with the number of clusters
kmeans = KMeans(n_clusters=3)

# fit the data to the Kmeans algorithm
kmeans.fit(X)

# get the predicted cluster labels for each data point
labels = kmeans.labels_
df['labels'] = labels
# plot the data points with different colors for the differnt clusters

plt.figure(figsize=(10,8))
plt.scatter(X.iloc[:,0],X.iloc[:,1],c=labels)
plt.xlabel('carat')
plt.ylabel('Price')
plt.title('K-Means Clustering on diamond dataset')
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('diamonds')

# generate sample data
X = df[['carat','price','cut','color','clarity']]
# Encode categorical feature 
X = pd.get_dummies(X)


# instantiate a Kmeans object with the number of clusters
kmeans = KMeans(n_clusters=5)

# fit the data to the Kmeans algorithm
kmeans.fit(X)

# get the predicted cluster labels for each data point
labels = kmeans.labels_
df['labels'] = labels
# plot the data points with different colors for the differnt clusters

plt.figure(figsize=(10,8))
plt.scatter(X.iloc[:,0],X.iloc[:,1],c=labels)
plt.xlabel('carat')
plt.ylabel('Price')
plt.title('K-Means Clustering on diamond dataset')
plt.show()

# plot the data points with different colors for different clusters

plt.figure(figsize=(10,8))
plt.scatter(X.iloc[:,2],X.iloc[:,1],c=labels)
plt.xlabel('cut')
plt.ylabel('Price')
plt.title('K-Means Clustering on diamond dataset')
plt.show()

plt.figure(figsize=(10,8))
plt.scatter(X.iloc[:,3],X.iloc[:,1],c=labels)
plt.xlabel('colur')
plt.ylabel('Price')
plt.title('K-Means Clustering on diamond dataset')
plt.show()

plt.figure(figsize=(10,8))
plt.scatter(X.iloc[:,4],X.iloc[:,1],c=labels)
plt.xlabel('clarity')
plt.ylabel('Price')
plt.title('K-Means Clustering on diamond dataset')
plt.show()
```
